```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Mengambil nama kolom (fitur) dari dataset
features = dataset.columns

# Membuat subplot untuk setiap fitur
fig, axes = plt.subplots(len(features), 1, figsize=(10, 4 * len(features)))

# Mengiterasi melalui setiap fitur dan membuat histogram
for i, feature in enumerate(features):
    sns.histplot(data=dataset, x=feature, hue='fetal_health', kde=True, ax=axes[i])
    axes[i].set_title(f'Distribusi Kelas untuk {feature}')

plt.tight_layout()  # Menyesuaikan tata letak subplot
plt.show()  # Menampilkan plot
```

3. Memvalidasi Data

```
import numpy as np

# 1. Memeriksa Missing Value (lebih detail)
# Menghitung jumlah missing value di setiap kolom
missing_values = dataset.isnull().sum()

# Menghitung persentase missing value di setiap kolom
missing_percentage = (missing_values / len(dataset)) * 100

# Membuat DataFrame untuk menampilkan missing value dan persentase
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
print("Missing Values:\n", missing_data)

# 2. Memeriksa Noisy Data (Outlier) dengan Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

# Fitur-fitur yang akan diperiksa outlier-nya
features_to_check = ['baseline value',  'accelerations', 'fetal_movement', 'uterine_contractions',
                     'light_decelerations','severe_decelerations', 'prolongued_decelerations',
                     'abnormal_short_term_variability', 'mean_value_of_short_term_variability',
                     'percentage_of_time_with_abnormal_long_term_variability',
                     'mean_value_of_long_term_variability', 'histogram_width',
                     'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
                     'histogram_number_of_zeroes', 'histogram_mode',
                     'histogram_mean', 'histogram_median', 'histogram_variance',
                     'histogram_tendency']

# Membuat boxplot untuk fitur-fitur yang dipilih
sns.boxplot(data=dataset[features_to_check])
plt.xticks(rotation=90)  # Rotasi label sumbu x agar terbaca
plt.title("Boxplot untuk Mendeteksi Outlier")
plt.show()

# 3. Memeriksa Data yang Tidak Sesuai (lebih spesifik)
# Contoh: Memeriksa apakah 'fetal_movement' > 1000 (tidak realistis)
invalid_fetal_movement = (dataset['fetal_movement'] > 1000).sum()
print("\nJumlah 'fetal_movement' > 1000:", invalid_fetal_movement)

# ... (lanjutkan pemeriksaan data yang tidak sesuai untuk fitur lain)

# 4. Memeriksa Konsistensi Data
# Contoh: Memeriksa apakah 'abnormal_short_term_variability' selalu
# lebih besar dari 'mean_value_of_short_term_variability'
inconsistent_data = dataset[dataset['abnormal_short_term_variability'] <
                                dataset['mean_value_of_short_term_variability']].shape[0]
print("\nJumlah Data Tidak Konsisten:", inconsistent_data)

# ... (lanjutkan pemeriksaan konsistensi untuk fitur lain)

# 5. Statistik Deskriptif untuk Memahami Distribusi Data
data_description = dataset.describe()
print("\nStatistik Deskriptif:\n", data_description)
```

4. Menentukan Objek Data

RINGKASAN
Objek Data:

baseline value: Nilai dasar denyut jantung janin (FHR) dalam denyut per menit (bpm).
accelerations: Jumlah akselerasi FHR per detik.
fetal_movement: Jumlah gerakan janin per detik.
uterine_contractions: Jumlah kontraksi uterus per detik.
light_decelerations: Jumlah deselerasi ringan FHR per detik.
severe_decelerations: Jumlah deselerasi berat FHR per detik.
prolongued_decelerations: Jumlah deselerasi berkepanjangan FHR per detik.
abnormal_short_term_variability: Persentase waktu dengan variabilitas jangka pendek yang abnormal.
mean_value_of_short_term_variability: Nilai rata-rata variabilitas jangka pendek.
percentage_of_time_with_abnormal_long_term_variability: Persentase waktu dengan variabilitas jangka panjang yang abnormal.
mean_value_of_long_term_variability: Nilai rata-rata variabilitas jangka panjang.
histogram_width: Lebar histogram FHR.
histogram_min: Nilai minimum histogram FHR.
histogram_max: Nilai maksimum histogram FHR.
histogram_number_of_peaks: Jumlah puncak dalam histogram FHR.
histogram_number_of_zero_crossings: Jumlah persilangan nol dalam histogram FHR.
histogram_mode: Modus histogram FHR.
histogram_mean: Rata-rata histogram FHR.
histogram_median: Median histogram FHR.
histogram_variance: Varians histogram FHR.
histogram_tendency: Kecenderungan histogram FHR (cth., datar, meningkat, menurun).
fetal_health: Klasifikasi kesehatan janin (1 = Normal, 2 = Suspect, 3 = Pathological).
Keterangan:

Objek data di atas merupakan fitur atau variabel yang diukur atau diamati dari data cardiotocography (CTG).
fetal_health adalah target variabel atau variabel dependen yang ingin diprediksi.
Fitur-fitur lainnya adalah variabel independen yang digunakan untuk memprediksi kesehatan janin.

```
# Mengubah tipe data 'fetal_health' menjadi object dan memetakan nilai
dataset['fetal_health'] = dataset['fetal_health'].map({
    1: 'Normal',
    2: 'Suspect',
    3: 'Pathological'
}).astype(str)  # Mengubah tipe data ke string (object)

# Menampilkan tipe data setelah diubah
print("\nTipe Data Setelah Diubah:\n", dataset.dtypes)

# Menampilkan distribusi kelas target setelah diubah
print("\nDistribusi Kelas Target (Setelah Diubah):\n", dataset['fetal_health'].value_counts())
```

5. Membersihkan Data

```
# menghitung nilai Null pada dataset
dataset.isnull().sum()
```

```
# mendeteksi keberadaan nilai Null
dataset.loc[:, dataset.isnull().any()].columns
```

```
# Mengubah Type data dari salah satu fitur
# Instead of directly converting to int, first map the string values back to their original numerical values
dataset['fetal_health'] = dataset['fetal_health'].map({
    'Normal': 1,
    'Suspect': 2,
    'Pathological': 3
}).astype(int) # Then convert to int
```

```
# Menampilkan Informasi dari data
dataset.info()
```

```
import missingno
```

```
# Memvisualisasikan keberadaan nilai Null
missingdata_df = dataset.columns[dataset.isnull().any()].tolist()
missingno.matrix(dataset[missingdata_df])
```

```
# karena tidak lebih dari 50% maka tidak perlu dihapus kolomnya
median_value=dataset['uterine_contractions'].median()
dataset['uterine_contractions']=dataset['uterine_contractions'].fillna(median_value)

median_value=dataset['prolongued_decelerations'].median()
dataset['prolongued_decelerations']=dataset['prolongued_decelerations'].fillna(median_value)

median_value=dataset['histogram_mode'].median()
dataset['histogram_mode']=dataset['histogram_mode'].fillna(median_value)

median_value=dataset['histogram_variance'].median()
dataset['histogram_variance']=dataset['histogram_variance'].fillna(median_value)

median_value=dataset['histogram_tendency'].median()
dataset['histogram_tendency']=dataset['histogram_tendency'].fillna(median_value)
```

## Ilmu data
</div>

Dengan Colab, Anda dapat memanfaatkan kecanggihan library Python yang populer untuk menganalisis dan memvisualisasikan data. Sel kode di bawah menggunakan <strong>numpy</strong> untuk menghasilkan beberapa data acak, dan menggunakan <strong>matplotlib</strong> untuk memvisualisasikannya. Untuk mengedit kode, cukup klik sel dan mulai pengeditan.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
